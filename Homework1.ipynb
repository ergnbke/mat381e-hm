```python
from urllib.request import urlopen
from xmltodict import parse
import pandas as pd
```

# Answer 1

```python
with urlopen("https://data.ibb.gov.tr/dataset/b47b5391-bcca-4bb3-a575-8ece68901d5d/resource/bcbeff5d-14d7-4ec0-a211-4c4d3e96cfba/download/istanbul-ehir-hatlar-iskeleleri.kml") as conn:
    raw = parse(conn.read().decode('utf8'))
raw
```

```python
rawdata = raw['kml']['Document']['Folder']['Folder'][0]['Placemark']

for i in range(1,5):
    rawdata.extend(raw['kml']['Document']['Folder']['Folder'][i]['Placemark'])
data = {x['name'] : x['Point']['coordinates'] for x in rawdata}
dataa =[]
for x in data:
    templon = data[x][:data[x].find(',')]
    templat =data[x][data[x].find(',')+1:data[x].find(',0')]
    dataa.append({'name' : x, 'longitude' :templon, 'latitude' : templat})

    
df = pd.DataFrame(dataa)
df.set_index('name')
```
